```python
checkpoint_path = "../data/checkpoints/subset_2_training_1/subset_cnn_sub_2-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

# Callback to save model's weights
cp_callback =tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    save_best_only=True,
    monitor='val_accuracy'
)
```

## Create Model (model_cnn_sub_2)

```python
model_cnn_sub_2 = tf.keras.Sequential()

model_cnn_sub_2.add(keras.layers.Conv1D(32, (36), 
                                        activation='relu', 
                                        input_shape=(129,376)))


model_cnn_sub_2.add(keras.layers.MaxPooling1D(36))

model_cnn_sub_2.add(keras.layers.Flatten())

# # dense layers...
model_cnn_sub_2.add(keras.layers.Dense(1024, activation='relu'))
model_cnn_sub_2.add(keras.layers.Dropout(0.1))
print(model_cnn_sub_2.output_shape)


model_cnn_sub_2.add(keras.layers.Dense(512, activation='relu'))
model_cnn_sub_2.add(keras.layers.Dropout(0.1))
print(model_cnn_sub_2.output_shape)

model_cnn_sub_2.add(keras.layers.Dense(64, activation='relu'))
model_cnn_sub_2.add(keras.layers.Dropout(0.1))
print(model_cnn_sub_2.output_shape)

model_cnn_sub_2.add(keras.layers.Dense(9, activation='softmax'))
print(model_cnn_sub_2.output_shape)

model_cnn_sub_2.summary()
```

```python
model_cnn_sub_2.compile(optimizer='Adam', 
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])
```

```python
cnn_sub_2_history = model_cnn_sub_2.fit(x=sub_X_train_np_arr, 
                                        y=sub_y_train,
#                                         validation_split=0.15,
                                        validation_data=(sub_X_test_np_arr, sub_y_test),
                                        callbacks=cp_callback,
                                        epochs=20)
```
